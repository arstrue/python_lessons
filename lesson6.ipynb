## 2. –†–∞–±–æ—Ç–∞ —Å –æ—Ç–¥–µ–ª—å–Ω—ã–º–∏ –±—É–∫–≤–∞–º–∏

‚ö†Ô∏è **–í–∞–∂–Ω–æ:** –°—Ç—Ä–æ–∫–∏ –Ω–µ–ª—å–∑—è –º–µ–Ω—è—Ç—å! –ï—Å–ª–∏ —Ö–æ—á–µ—à—å –∏–∑–º–µ–Ω–∏—Ç—å —Ç–µ–∫—Å—Ç ‚Äî –Ω—É–∂–Ω–æ —Å–æ–∑–¥–∞—Ç—å –Ω–æ–≤—ã–π:

```python
text = "–∫–æ—Ç"
# text[0] = "–ö"  ‚Üê –¢–∞–∫ –Ω–µ–ª—å–∑—è! –ë—É–¥–µ—Ç –æ—à–∏–±–∫–∞ ‚ùå

# –ü—Ä–∞–≤–∏–ª—å–Ω–æ ‚Äî —Å–æ–∑–¥–∞—ë–º –Ω–æ–≤—ã–π —Ç–µ–∫—Å—Ç:
text = "–ö" + text[1:]
print(text)  # ‚Üí –ö–æ—Ç ‚úÖ
```

## –ü–∞–º—è—Ç–∫–∞ –ø–æ —Å—Ç—Ä–æ–∫–∞–º

| –ß—Ç–æ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å | –ö–∞–∫ –Ω–∞–ø–∏—Å–∞—Ç—å |

| –£–∑–Ω–∞—Ç—å –¥–ª–∏–Ω—É —Å—Ç—Ä–æ–∫–∏ | `len(text)` |

## 3. –í–æ–ª—à–µ–±–Ω—ã–µ –∫–æ–º–∞–Ω–¥—ã –¥–ª—è —Å—Ç—Ä–æ–∫ (–º–µ—Ç–æ–¥—ã)

–£ —Å—Ç—Ä–æ–∫ –µ—Å—Ç—å –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã–µ –ø–æ–º–æ—â–Ω–∏–∫–∏ ‚Äî –º–µ—Ç–æ–¥—ã. –î–∞–≤–∞–π –ø–æ–∑–Ω–∞–∫–æ–º–∏–º—Å—è —Å —Å–∞–º—ã–º–∏ –ø–æ–ª–µ–∑–Ω—ã–º–∏:

```python
text = "–ø—Ä–∏–≤–µ—Ç, –º–∏—Ä!"

# –î–µ–ª–∞–µ–º –≤—Å–µ –±—É–∫–≤—ã –±–æ–ª—å—à–∏–º–∏
print(text.upper())  # ‚Üí –ü–†–ò–í–ï–¢, –ú–ò–†!

# –î–µ–ª–∞–µ–º –≤—Å–µ –±—É–∫–≤—ã –º–∞–ª–µ–Ω—å–∫–∏–º–∏
print(text.lower())  # ‚Üí –ø—Ä–∏–≤–µ—Ç, –º–∏—Ä!

# –ü–µ—Ä–≤–∞—è –±—É–∫–≤–∞ –±–æ–ª—å—à–∞—è
print(text.capitalize())  # ‚Üí –ü—Ä–∏–≤–µ—Ç, –º–∏—Ä!

# –£–±–∏—Ä–∞–µ–º –ª–∏—à–Ω–∏–µ –ø—Ä–æ–±–µ–ª—ã –ø–æ –∫—Ä–∞—è–º
spaces = "   –ø—Ä–∏–≤–µ—Ç   "
print(spaces.strip())  # ‚Üí "–ø—Ä–∏–≤–µ—Ç"
```

## 4. –ï—â—ë –±–æ–ª—å—à–µ –ø–æ–ª–µ–∑–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤

```python
sentence = "–Ø –ª—é–±–ª—é –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞—Ç—å"

# –†–∞–∑–±–∏–≤–∞–µ–º —Ç–µ–∫—Å—Ç –Ω–∞ —Å–ª–æ–≤–∞
words = sentence.split()
print(words)  # ‚Üí ['–Ø', '–ª—é–±–ª—é', '–ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞—Ç—å']

# –°—á–∏—Ç–∞–µ–º, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –±—É–∫–≤–∞
print(sentence.count("—é"))  # ‚Üí 2

# –ó–∞–º–µ–Ω—è–µ–º —Å–ª–æ–≤–æ –Ω–∞ –¥—Ä—É–≥–æ–µ
new_sentence = sentence.replace("–ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞—Ç—å", "—Ä–∏—Å–æ–≤–∞—Ç—å")
print(new_sentence)  # ‚Üí –Ø –ª—é–±–ª—é —Ä–∏—Å–æ–≤–∞—Ç—å

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è –ª–∏ —Å—Ç—Ä–æ–∫–∞ —Å –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–π –±—É–∫–≤—ã
print(sentence.startswith("–Ø"))  # ‚Üí True

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, –∑–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç—Å—è –ª–∏ —Å—Ç—Ä–æ–∫–∞ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ–π –±—É–∫–≤–æ–π
print(sentence.endswith("—å"))  # ‚Üí True
```
