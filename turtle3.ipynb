–ï—Å–ª–∏ –≤–Ω—É—Ç—Ä–∏ `func` —Å–Ω–æ–≤–∞ –≤—ã–∑–≤–∞—Ç—å `ontimer(func, delay)`, –ø–æ–ª—É—á–∏—Ç—Å—è **–±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–π —Ü–∏–∫–ª –±–µ–∑ –∑–∞–≤–∏—Å–∞–Ω–∏—è**!

#### –ü—Ä–∏–º–µ—Ä: –í—Ä–∞—â–∞—é—â–∏–π—Å—è –∫–≤–∞–¥—Ä–∞—Ç

````python
import turtle

screen = turtle.Screen()
t = turtle.Turtle()
t.shape("square")
t.penup()

def animate():
    t.right(5)          # –ø–æ–≤–æ—Ä–∞—á–∏–≤–∞–µ–º
    screen.ontimer(animate, 50)  # –ø–æ–≤—Ç–æ—Ä—è–µ–º —á–µ—Ä–µ–∑ 50 –º—Å

animate()               # –∑–∞–ø—É—Å–∫
screen.mainloop()       # –∂–¥—ë–º –∑–∞–∫—Ä—ã—Ç–∏—è –æ–∫–Ω–∞
````

3. –í–∫–ª—é—á–∏—Ç—å –ø—Ä–æ—Å–ª—É—à–∏–≤–∞–Ω–∏–µ: `screen.listen()`.

````python
import turtle

screen = turtle.Screen()
t = turtle.Turtle()

def go_forward():
    t.forward(20)

def turn_left():
    t.left(15)

def turn_right():
    t.right(15)

# –ü—Ä–∏–≤—è–∑–∫–∞ –∫–ª–∞–≤–∏—à
screen.onkey(go_forward, "Up")
screen.onkey(turn_left, "Left")
screen.onkey(turn_right, "Right")

screen.listen()  # –í–∫–ª—é—á–∏—Ç—å –ø—Ä–æ—Å–ª—É—à–∏–≤–∞–Ω–∏–µ!
screen.mainloop()
````
