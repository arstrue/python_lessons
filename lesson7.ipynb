## 1. Зачем нужны файлы?

Пока что все наши программы хранили данные только в памяти — как только программа заканчивалась, всё исчезало.

**Файлы** позволяют сохранять информацию надолго — как записная книжка или дневник. Ты можешь:
- Сохранить список дел
- Записать результаты вычислений
- Прочитать целую книгу прямо в программе!

Всё это делается с помощью команды `open()`.

## 1.1. Пути к файлам
В будущем на уроках информатики вам расскажут, что такое абсолютные и относительные пути к файлам, и вы сможете обращаться из своей программы к любым файлам на компьютере, где бы они ни находились.
Сейчас же нам это не очень важно, поэтому мы будем работать только с файлами, которые находятся **в одной папке** с программой. К таким файлам можно обращаться просто по имени, python вас поймет и найдет нужный файл.

## 2. Как записать в файл?

Чтобы записать текст в файл, используем режим `'w'` (write = писать).

````python
# Создаём файл и записываем в него текст
with open("приветствие.txt", "w") as f:
    f.write("Привет! Это мой первый файл в Python.\n")
    f.write("Я могу писать сколько угодно строк! 🌟")

print("✅ Файл 'приветствие.txt' создан!")
````

💡 **Используй параметр `encoding="utf-8"` функции open(), если русские буквы сохранились неправильно**

## 3. Как прочитать файл?

Чтобы прочитать содержимое файла, используем режим `'r'` (read = читать).

````python
# Сначала создадим файл для примера
with open("заметки.txt", "w", encoding="utf-8") as f:
    f.write("1. Покормить кота\n")
    f.write("2. Сделать уроки\n")
    f.write("3. Поиграть в футбол\n")

# Теперь прочитаем его
with open("заметки.txt", "r", encoding="utf-8") as f:
    content = f.read()

print("Содержимое файла:\n")
print(content)
````

## 4. Режимы работы с файлами

| Режим | Назначение |
|-------|------------|
| `'r'` | Чтение (файл должен существовать) |
| `'w'` | Запись (если файл есть — сотрётся!) |
| `'a'` | Добавление в конец файла (не стирает старое) |
| `'r+'`| Чтение и запись |

✅ **Всегда используй конструкцию `with open(...) as f:`** — она сама закроет файл, даже если программа сломается!

## 5. Чтение построчно

Иногда файл большой, и читать его целиком неудобно. Тогда можно перебирать строки:

````python
with open("заметки.txt", "r", encoding="utf-8") as f:
    for line in f:
        print("✅", line.strip())  # strip() убирает \n в конце строки
````
