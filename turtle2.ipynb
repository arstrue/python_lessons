## 1. –¶–∏–∫–ª `for` ‚Äî —Ç–≤–æ–π —Ä–æ–±–æ—Ç-–ø–æ–≤—Ç–æ—Ä—è–ª—å—â–∏–∫

–•–æ—á–µ—à—å –ø–æ–≤—Ç–æ—Ä–∏—Ç—å –∫–æ–º–∞–Ω–¥—É 4 —Ä–∞–∑–∞? –ü–∏—à–∏:
```python
for i in range(4):
    t.forward(100)
    t.left(90)
```

‚úÖ `for i in range(4):` ‚Äî –∑–Ω–∞—á–∏—Ç *¬´—Å–¥–µ–ª–∞–π 4 —Ä–∞–∑–∞¬ª*
- –í—Å—ë, —á—Ç–æ —Å **–æ—Ç—Å—Ç—É–ø–æ–º**, –ø–æ–≤—Ç–æ—Ä–∏—Ç—Å—è 4 —Ä–∞–∑–∞.
- `i` ‚Äî —Å—á—ë—Ç—á–∏–∫ (–º–æ–∂–Ω–æ –Ω–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å, –ø—Ä–æ—Å—Ç–æ –Ω—É–∂–µ–Ω –ø–æ –ø—Ä–∞–≤–∏–ª–∞–º Python).

````python
import turtle as t

for i in range(4):
    t.forward(100)
    t.left(90)

t.done()
````

üí° –≠—Ç–æ —Ç–æ—Ç –∂–µ –∫–≤–∞–¥—Ä–∞—Ç, –Ω–æ –≤ 2 —Ä–∞–∑–∞ –∫–æ—Ä–æ—á–µ!

–ê —Ç–µ–ø–µ—Ä—å ‚Äî –Ω–∞—Å—Ç–æ—è—â–∞—è –º–∞–≥–∏—è.

## 2. –ú–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∏ –∑–∞ 1 –º–∏–Ω—É—Ç—É

–•–æ—á–µ—à—å **–ø—è—Ç–∏—É–≥–æ–ª—å–Ω–∏–∫**? –£–≥–æ–ª = 360¬∞ / 5 = 72¬∞.

–®–µ—Å—Ç–∏—É–≥–æ–ª—å–Ω–∏–∫? 360¬∞ / 6 = 60¬∞.

–§–æ—Ä–º—É–ª–∞: **`360 / –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ_—Å—Ç–æ—Ä–æ–Ω`**

````python
import turtle as t

t.color("purple")
ugol = 360 / 6  # —à–µ—Å—Ç–∏—É–≥–æ–ª—å–Ω–∏–∫

for i in range(6):
    t.forward(80)
    t.left(ugol)

t.done()
````

üëâ –ü–æ–ø—Ä–æ–±—É–π –∏–∑–º–µ–Ω–∏—Ç—å `6` –Ω–∞ `8`, `12`, `20` ‚Äî –∏ –ø–æ—Å–º–æ—Ç—Ä–∏, —á—Ç–æ –±—É–¥–µ—Ç!

## 3. –°–ø–∏—Ä–∞–ª—å üåÄ

–°–ø–∏—Ä–∞–ª—å ‚Äî —ç—Ç–æ –∫–æ–≥–¥–∞ –∫–∞–∂–¥–∞—è –ª–∏–Ω–∏—è —á—É—Ç—å –¥–ª–∏–Ω–Ω–µ–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–π.

–ò—Å–ø–æ–ª—å–∑—É–µ–º `i` –∏–∑ —Ü–∏–∫–ª–∞ ‚Äî –æ–Ω —Ä–∞—Å—Ç—ë—Ç: 0, 1, 2, 3‚Ä¶

````python
import turtle as t

t.speed(0)  # –±—ã—Å—Ç—Ä–æ!

for i in range(60):
    t.forward(i * 2)  # 0, 2, 4, 6, 8...
    t.left(30)        # –ø–æ–≤–æ—Ä–æ—Ç –Ω–∞ 30¬∞

t.done()
````

üß† –ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç `i * 2`?
- –ù–∞ 1-–º —à–∞–≥–µ: `i = 0` ‚Üí `forward(0)`
- –ù–∞ 2-–º: `i = 1` ‚Üí `forward(2)`
- –ù–∞ 10-–º: `i = 9` ‚Üí `forward(18)`
- –ò —Ç–∞–∫ ‚Äî –ª–∏–Ω–∏–∏ –≤—Å—ë –¥–ª–∏–Ω–Ω–µ–µ ‚Üí —Å–ø–∏—Ä–∞–ª—å!

## 4. –¶–≤–µ—Ç–Ω–∞—è —Å–ø–∏—Ä–∞–ª—å üåà

–î–æ–±–∞–≤–∏–º —Ü–≤–µ—Ç! –í `turtle` –µ—Å—Ç—å –≥–æ—Ç–æ–≤—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è:

```python
colors = ["red", "orange", "yellow", "lime", "blue", "purple"]
```

–ê `i % 6` ‚Äî –¥–∞—ë—Ç –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è: 0,1,2,3,4,5,0,1,2‚Ä¶ ‚Äî –∏–¥–µ–∞–ª—å–Ω–æ –¥–ª—è —Ü–∏–∫–ª–∏—á–µ—Å–∫–æ–π —Å–º–µ–Ω—ã —Ü–≤–µ—Ç–∞!

````python
import turtle as t

t.speed(0)
colors = ["red", "orange", "yellow", "lime", "blue", "purple"]

for i in range(120):
    t.color(colors[i % 6])   # –≤—ã–±–∏—Ä–∞–µ–º —Ü–≤–µ—Ç –ø–æ –Ω–æ–º–µ—Ä—É
    t.forward(i * 2)
    t.left(59)               # 59¬∞ ‚Äî –ø–æ—á—Ç–∏ 60, –Ω–æ –Ω–µ —Å–æ–≤—Å–µ–º ‚Üí –∫—Ä–∞—Å–∏–≤–∞—è —Å–ø–∏—Ä–∞–ª—å!

t.done()
````

üé® –ü–æ–ø—Ä–æ–±—É–π –∏–∑–º–µ–Ω–∏—Ç—å `59` –Ω–∞ `91`, `121`, `179` ‚Äî –ø–æ–ª—É—á–∞—é—Ç—Å—è —Ä–∞–∑–Ω—ã–µ —É–∑–æ—Ä—ã!

## 5. –°–Ω–µ–∂–∏–Ω–∫–∞ –∏–∑ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤ ‚ùÑÔ∏è

–ò–¥–µ—è: –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ ‚Üí –ø–æ–≤–µ—Ä–Ω—É—Ç—å ‚Üí –ø–æ–≤—Ç–æ—Ä–∏—Ç—å 6 —Ä–∞–∑.

–ë—É–¥–µ—Ç –∫–∞–∫ —Å–Ω–µ–∂–∏–Ω–∫–∞!

````python
import turtle as t

t.speed(0)
t.pensize(2)

for i in range(6):
    # –†–∏—Å—É–µ–º –æ–¥–∏–Ω —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫
    for j in range(3):
        t.forward(60)
        t.left(120)
    # –ü–æ–≤–æ—Ä–∞—á–∏–≤–∞–µ–º—Å—è –∫ —Å–ª–µ–¥—É—é—â–µ–º—É –ª—É—á—É
    t.left(60)

t.done()
````

üîç –ó–¥–µ—Å—å **–¥–≤–∞ —Ü–∏–∫–ª–∞**: –≤–Ω–µ—à–Ω–∏–π ‚Äî 6 —Ä–∞–∑ (–ª—É—á–∏), –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π ‚Äî 3 —Ä–∞–∑–∞ (—Å—Ç–æ—Ä–æ–Ω—ã —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞).

–≠—Ç–æ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è **–≤–ª–æ–∂–µ–Ω–Ω—ã–π —Ü–∏–∫–ª** ‚Äî –∫–∞–∫ –º–∞—Ç—Ä—ë—à–∫–∞!

## 6. –ß–µ—Ä–µ–ø–∞—à–∫–∞-—Ç–∞–Ω—Ü–æ—Ä üï∫

–ê —Ç–µ–ø–µ—Ä—å ‚Äî –∞–Ω–∏–º–∞—Ü–∏—è! –ß–µ—Ä–µ–ø–∞—à–∫–∞ –±—É–¥–µ—Ç ¬´—Ç–∞–Ω—Ü–µ–≤–∞—Ç—å¬ª:
- –ö—Ä—É–∂–∏—Ç—å—Å—è,
- –ú–µ–Ω—è—Ç—å —Ü–≤–µ—Ç,
- –û—Å—Ç–∞–≤–ª—è—Ç—å —Å–ª–µ–¥.

(–í Jupyter –∞–Ω–∏–º–∞—Ü–∏—è –º–æ–∂–µ—Ç –Ω–µ –ø–ª–∞–≤–∞—Ç—å ‚Äî –ª—É—á—à–µ –∑–∞–ø—É—Å–∫–∞—Ç—å –≤ Thonny/IDLE.)

````python
import turtle as t

t.speed(0)
t.shape("turtle")
t.pensize(3)

colors = ["red", "orange", "yellow", "green"]

for i in range(48):
    t.color(colors[i % 4])
    t.forward(100)
    t.backward(100)
    t.right(7.5)  # 360 / 48 = 7.5¬∞ ‚Äî –ø–æ–ª–Ω—ã–π –∫—Ä—É–≥ –∑–∞ 48 —à–∞–≥–æ–≤

t.done()
````
