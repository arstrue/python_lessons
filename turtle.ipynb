## 1. –ü–æ–¥–∫–ª—é—á–∞–µ–º —á–µ—Ä–µ–ø–∞—à–∫—É ‚Äî –∫–æ—Ä–æ—Ç–∫–æ!

–í–º–µ—Å—Ç–æ –¥–ª–∏–Ω–Ω–æ–≥–æ `turtle.forward(100)` –º—ã –±—É–¥–µ–º –ø–∏—Å–∞—Ç—å –ø—Ä–æ—Å—Ç–æ `t.forward(100)`.

–î–µ–ª–∞–µ—Ç—Å—è —ç—Ç–æ –æ–¥–Ω–æ–π —Å—Ç—Ä–æ—á–∫–æ–π:

````python
import turtle as t
````

‚úÖ `as t` ‚Äî –∑–Ω–∞—á–∏—Ç: *¬´–Ω–∞–∑–æ–≤–∏ –±–∏–±–ª–∏–æ—Ç–µ–∫—É `turtle` –∫–æ—Ä–æ—Ç–∫–∏–º –∏–º–µ–Ω–µ–º `t`¬ª*.

–¢–µ–ø–µ—Ä—å –≤—Å–µ –∫–æ–º–∞–Ω–¥—ã –Ω–∞—á–∏–Ω–∞—é—Ç—Å—è —Å `t.` ‚Äî –∏ —ç—Ç–æ –±—ã—Å—Ç—Ä–æ!

## 2. –ü—Ä–æ—Å—Ç–µ–π—à–∏–µ –∫–æ–º–∞–Ω–¥—ã

–ß–µ—Ä–µ–ø–∞—à–∫–∞ –ø–æ–Ω–∏–º–∞–µ—Ç —Ç–∞–∫–∏–µ –∫–æ–º–∞–Ω–¥—ã:

| –ö–æ–º–∞–Ω–¥–∞ | –ß—Ç–æ –¥–µ–ª–∞–µ—Ç |
|--------|-------------|
| `t.forward(100)` | –ü–æ–ª–∑—ë—Ç **–≤–ø–µ—Ä—ë–¥** –Ω–∞ 100 —à–∞–≥–æ–≤ |
| `t.backward(50)` | –ü–æ–ª–∑—ë—Ç **–Ω–∞–∑–∞–¥** –Ω–∞ 50 —à–∞–≥–æ–≤ |
| `t.left(90)` | –ü–æ–≤–æ—Ä–∞—á–∏–≤–∞–µ—Ç—Å—è **–≤–ª–µ–≤–æ** –Ω–∞ 90¬∞ |
| `t.right(45)` | –ü–æ–≤–æ—Ä–∞—á–∏–≤–∞–µ—Ç—Å—è **–≤–ø—Ä–∞–≤–æ** –Ω–∞ 45¬∞ |

–î–∞–≤–∞–π –Ω–∞—Ä–∏—Å—É–µ–º **–∫–≤–∞–¥—Ä–∞—Ç** ‚Äî –≤—Å–µ–≥–æ 5 —Å—Ç—Ä–æ—á–µ–∫!

````python
import turtle as t

t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)

t.done()  # —á—Ç–æ–±—ã –æ–∫–Ω–æ –Ω–µ –∑–∞–∫—Ä—ã–ª–æ—Å—å
````

‚úÖ –°–æ–≤–µ—Ç:
- –ó–∞–ø—É—Å—Ç–∏ —è—á–µ–π–∫—É ‚Äî –ø–æ—è–≤–∏—Ç—Å—è –æ–∫–Ω–æ —Å —á–µ—Ä–µ–ø–∞—à–∫–æ–π –∏ –∫–≤–∞–¥—Ä–∞—Ç–æ–º!
- `t.done()` –≤ –∫–æ–Ω—Ü–µ ‚Äî —á—Ç–æ–±—ã –æ–∫–Ω–æ –æ—Å—Ç–∞–ª–æ—Å—å –æ—Ç–∫—Ä—ã—Ç—ã–º.

## 3. –¶–≤–µ—Ç –∏ —Ç–æ–ª—â–∏–Ω–∞ –ª–∏–Ω–∏–∏

–ú–æ–∂–Ω–æ –º–µ–Ω—è—Ç—å —Ü–≤–µ—Ç –∏ —Ç–æ–ª—â–∏–Ω—É –ª–∏–Ω–∏–∏ ‚Äî —Ç–æ–∂–µ —á–µ—Ä–µ–∑ `t.`:

````python
import turtle as t

t.color("red")      # —Ü–≤–µ—Ç –ø–µ—Ä–∞ ‚Äî –∫—Ä–∞—Å–Ω—ã–π
t.pensize(5)         # —Ç–æ–ª—â–∏–Ω–∞ –ª–∏–Ω–∏–∏ ‚Äî 5

t.forward(150)
t.left(120)
t.forward(150)
t.left(120)
t.forward(150)

t.done()
````

üé® –¶–≤–µ—Ç–∞ –º–æ–∂–Ω–æ –ø–∏—Å–∞—Ç—å –ø–æ-–∞–Ω–≥–ª–∏–π—Å–∫–∏: `"blue"`, `"green"`, `"purple"`, `"orange"`, `"pink"`, `"brown"`, `"gold"`, `"lime"`‚Ä¶

–ü–æ–ø—Ä–æ–±—É–π —Ä–∞–∑–Ω—ã–µ ‚Äî –∏ –ø–æ—Å–º–æ—Ç—Ä–∏, —á—Ç–æ –ø–æ–ª—É—á–∏—Ç—Å—è!

## 4. –ü–æ–¥–Ω—è—Ç—å –∏ –æ–ø—É—Å—Ç–∏—Ç—å –ø–µ—Ä–æ

–ò–Ω–æ–≥–¥–∞ –Ω—É–∂–Ω–æ **–ø–µ—Ä–µ–º–µ—Å—Ç–∏—Ç—å—Å—è**, –Ω–æ **–Ω–µ —Ä–∏—Å–æ–≤–∞—Ç—å**.

–ö–æ–º–∞–Ω–¥—ã:
- `t.penup()` ‚Äî –ø–æ–¥–Ω—è—Ç—å –ø–µ—Ä–æ (–Ω–µ —Ä–∏—Å—É–µ—Ç)
- `t.pendown()` ‚Äî –æ–ø—É—Å—Ç–∏—Ç—å –ø–µ—Ä–æ (—Å–Ω–æ–≤–∞ —Ä–∏—Å—É–µ—Ç)

````python
import turtle as t

t.forward(50)

t.penup()      # –ø–æ–¥–Ω—è–ª–∏ –ø–µ—Ä–æ
t.forward(30)  # –ø–µ—Ä–µ—à–ª–∏ ‚Äî –ª–∏–Ω–∏–∏ –Ω–µ—Ç!
t.pendown()    # –æ–ø—É—Å—Ç–∏–ª–∏ –ø–µ—Ä–æ

t.forward(50)

t.done()
````

–ü–æ–ª—É—á–∏—Ç—Å—è: `____   ____` ‚Äî –¥–≤–µ –ª–∏–Ω–∏–∏ —Å –ø—Ä–æ–±–µ–ª–æ–º!

## 5. –ö—Ä—É–≥–∏ –∏ —Ç–æ—á–∫–∏

–ï—â—ë –ø–æ–ª–µ–∑–Ω—ã–µ –∫–æ–º–∞–Ω–¥—ã:

````python
import turtle as t

t.circle(50)          # –∫—Ä—É–≥ —Ä–∞–¥–∏—É—Å–æ–º 50

t.penup()
t.forward(100)
t.pendown()

t.dot(20, "red")      # —Ç–æ—á–∫–∞: –¥–∏–∞–º–µ—Ç—Ä 20, —Ü–≤–µ—Ç ‚Äî –∫—Ä–∞—Å–Ω—ã–π

t.done()
````

## 6. –°–æ–≤–µ—Ç—ã –¥–ª—è –∫–æ–º—Ñ–æ—Ä—Ç–Ω–æ–≥–æ —Ä–∏—Å–æ–≤–∞–Ω–∏—è

–î–æ–±–∞–≤—å –≤ –Ω–∞—á–∞–ª–æ (–ø–æ—Å–ª–µ `import`) —ç—Ç–∏ —Å—Ç—Ä–æ—á–∫–∏ ‚Äî –∏ —Ä–∏—Å–æ–≤–∞—Ç—å —Å—Ç–∞–Ω–µ—Ç —É–¥–æ–±–Ω–µ–µ:
