## 3. –ü—Ä–æ—Å—Ç–µ–π—à–∏–µ –∫–æ–º–∞–Ω–¥—ã

|--------|-------------|
| `t.forward(100)` | –ü–æ–ª–∑—ë—Ç **–≤–ø–µ—Ä—ë–¥** –Ω–∞ 100 —à–∞–≥–æ–≤ |

| `t.left(90)` | –ü–æ–≤–æ—Ä–∞—á–∏–≤–∞–µ—Ç—Å—è **–≤–ª–µ–≤–æ** –Ω–∞ 90¬∞ |

## 4. –¶–≤–µ—Ç –∏ —Ç–æ–ª—â–∏–Ω–∞ –ª–∏–Ω–∏–∏

–ú–æ–∂–Ω–æ –º–µ–Ω—è—Ç—å —Ü–≤–µ—Ç –ø–µ—Ä–∞ –∏ —Ç–æ–ª—â–∏–Ω—É –ª–∏–Ω–∏–∏:

````python
import turtle
t = turtle.Turtle()

t.color("red")        # —Ü–≤–µ—Ç –ø–µ—Ä–∞ ‚Äî –∫—Ä–∞—Å–Ω—ã–π
t.pensize(5)          # —Ç–æ–ª—â–∏–Ω–∞ –ª–∏–Ω–∏–∏ ‚Äî 5 –ø–∏–∫—Å–µ–ª–µ–π

t.forward(150)
t.left(120)
t.forward(150)
t.left(120)
t.forward(150)
````

## 5. –ü–æ–¥–Ω—è—Ç—å –∏ –æ–ø—É—Å—Ç–∏—Ç—å –ø–µ—Ä–æ

–ò–Ω–æ–≥–¥–∞ –Ω—É–∂–Ω–æ **–ø–µ—Ä–µ–º–µ—Å—Ç–∏—Ç—å—Å—è**, –Ω–æ **–Ω–µ —Ä–∏—Å–æ–≤–∞—Ç—å** ‚Äî –Ω–∞–ø—Ä–∏–º–µ—Ä, –ø–µ—Ä–µ–π—Ç–∏ –∫ –Ω–æ–≤–æ–º—É –º–µ—Å—Ç—É.

–î–ª—è —ç—Ç–æ–≥–æ –µ—Å—Ç—å –∫–æ–º–∞–Ω–¥—ã:
- `t.penup()` ‚Äî –ø–æ–¥–Ω—è—Ç—å –ø–µ—Ä–æ (–Ω–µ —Ä–∏—Å—É–µ—Ç)
- `t.pendown()` ‚Äî –æ–ø—É—Å—Ç–∏—Ç—å –ø–µ—Ä–æ (—Å–Ω–æ–≤–∞ —Ä–∏—Å—É–µ—Ç)

````python
import turtle
t = turtle.Turtle()

t.color("blue")
t.forward(50)

t.penup()      # –ø–æ–¥–Ω—è–ª–∏ –ø–µ—Ä–æ
t.forward(30)  # –ø–µ—Ä–µ—à–ª–∏ –≤–ø–µ—Ä—ë–¥ ‚Äî –ª–∏–Ω–∏–∏ –Ω–µ—Ç!
t.pendown()    # –æ–ø—É—Å—Ç–∏–ª–∏ –ø–µ—Ä–æ

t.forward(50)  # —Å–Ω–æ–≤–∞ —Ä–∏—Å—É–µ–º
````
