## 3. –ü—Ä–∏–º–µ—Ä: –ó–∞–∫—Ä–∞—à–µ–Ω–Ω—ã–π –∫–≤–∞–¥—Ä–∞—Ç üü•

–ù–∞—Ä–∏—Å—É–µ–º –∫–≤–∞–¥—Ä–∞—Ç –∏ –∑–∞–ª—å—ë–º –µ–≥–æ –∂—ë–ª—Ç—ã–º:

````python
import turtle as t

t.color("yellow")     # —Å–Ω–∞—á–∞–ª–∞ ‚Äî —Ü–≤–µ—Ç –∑–∞–ª–∏–≤–∫–∏ –∏ –∫–æ–Ω—Ç—É—Ä–∞
t.begin_fill()         # ‚Üê –Ω–∞—á–∞–ª–∏ –∑–∞–ª–∏–≤–∫—É

t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)

t.end_fill()           # ‚Üê –∑–∞–∫–æ–Ω—á–∏–ª–∏ –∑–∞–ª–∏–≤–∫—É

t.done()
````
